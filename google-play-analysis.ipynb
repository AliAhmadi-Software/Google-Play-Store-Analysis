## Strorytelling - Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'].dropna(), kde=True, bins=20)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Installs', y='Reviews', hue='Category', alpha=0.6)
plt.title('Number of Installs vs Reviews')
plt.xlabel('Installs')
plt.ylabel('Reviews')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
category_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
sns.barplot(x=category_rating, y=category_rating.index, palette='coolwarm')
plt.title('Average Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```

```python
sns.catplot(data=df, x='Content Rating', y='Rating', kind='bar', aspect=2, height=6)
plt.title('Average Rating per Content Rating Category')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(data=df, x='Installs', y='Rating', kind='violin', aspect=2, height=6)
plt.title('Distribution of Ratings across Install Bins')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(data=df, x='Category', y='Reviews', kind='box', aspect=2, height=6)
plt.title('Distribution of Reviews per Category')
plt.xticks(rotation=90)
plt.yscale('log')  # Scale the y-axis because of large range in review counts
plt.show()
```

```python
sns.catplot(data=df, x='Category', hue='Type', kind='count', aspect=2, height=6)
plt.title('Number of Free vs Paid Apps in Each Category')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.relplot(x='Reviews', y='Rating', hue='Category', size='Installs', data=df, sizes=(40, 400))
plt.title('Rating vs. Reviews per Category')
plt.xscale('log')
plt.show()
```

```python
sns.relplot(x='Size', y='Installs', hue='Content Rating', data=df, aspect=1.5)
sns.set_theme(style="darkgrid")
plt.title('Installs vs. Size per Content Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.relplot(x='Reviews', y='Rating', hue='Category', kind='line', data=df, height=6, aspect=1.5)
plt.title('Rating vs. Reviews per Category')
plt.xscale('log')
plt.show()
```

```python
sns.relplot(x='Size', y='Installs', hue='Content Rating', kind='line', data=df, height=6, aspect=1.5)
plt.title('Installs vs. Size per Content Rating')
plt.show()
```

```python
# ایجاد گروه‌های نصب
bins = [0, 10000, 100000, 1000000, 10000000, 100000000, 1000000000]
labels = ['0-10K', '10K-100K', '100K-1M', '1M-10M', '10M-100M', '100M+']
df['Install Groups'] = pd.cut(df['Installs'], bins=bins, labels=labels, include_lowest=True)

# نمایش تعداد هر گروه
install_group_counts = df['Install Groups'].value_counts().sort_index()
install_group_counts
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Install Groups', data=df, palette='viridis')
plt.title('Distribution of Apps by Install Groups')
plt.xlabel('Install Groups')
plt.ylabel('Count of Apps')
plt.show()
```

```python
# میانگین رتبه‌ها در هر گروه نصب
install_group_ratings = df.groupby('Install Groups')['Rating'].mean()

# رسم نمودار برای میانگین رتبه‌ها در گروه‌های نصب
plt.figure(figsize=(10, 6))
install_group_ratings.plot(kind='bar', color='skyblue')
plt.title('Average Rating by Install Groups')
plt.xlabel('Install Groups')
plt.ylabel('Average Rating')
plt.show()
```
